```python
# Import necessary libraries
import numpy as np
import pandas as pd
from scipy import stats

# Set random seed for reproducibility
np.random.seed(42)

# Create a sample dataset
# Let's assume we have three groups (A, B, C) and their respective observations
group_A = np.random.normal(loc=20, scale=5, size=30)
group_B = np.random.normal(loc=22, scale=5, size=30)
group_C = np.random.normal(loc=25, scale=5, size=30)

# Combine the groups into a single dataset
data = pd.DataFrame({
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30,
    'Value': np.concatenate([group_A, group_B, group_C])
})

# Display the first few rows of the dataset
print(data.head())

# Perform t-tests between pairs of groups
t_statistic_AB, p_value_AB = stats.ttest_ind(group_A, group_B)
t_statistic_AC, p_value_AC = stats.ttest_ind(group_A, group_C)
t_statistic_BC, p_value_BC = stats.ttest_ind(group_B, group_C)

print("T-Test Results:")
print("Group A vs Group B: t-statistic = {:.4f}, p-value = {:.4f}".format(t_statistic_AB, p_value_AB))
print("Group A vs Group C: t-statistic = {:.4f}, p-value = {:.4f}".format(t_statistic_AC, p_value_AC))
print("Group B vs Group C: t-statistic = {:.4f}, p-value = {:.4f}".format(t_statistic_BC, p_value_BC))

# Perform ANOVA test
f_statistic, p_value_anova = stats.f_oneway(group_A, group_B, group_C)
print("\nANOVA Test Results:")
print("F-statistic = {:.4f}, p-value = {:.4f}".format(f_statistic, p_value_anova))
```
